```python
import pandas as pd
import numpy as np
import matplotlib
import plotly.express as px  # (version 4.7.0 or higher)
import plotly.graph_objects as go
from dash import Dash, dcc, html, Input, Output 
```

```python
df_all = pd.read_csv('cdr_data-2.csv',parse_dates=['timestamp'])
```

```python
display(df_all[df_all.type=='message'].sort_values(by=['x_id','y_id'], ascending= True))
```

```python
df_all.dtypes
```

```python
df= df_all[df_all.type=='call']
top_calls_22 = df[['y_id','length']][df.x_id==22].groupby(['y_id']).sum().sort_values(by='length',ascending=False)
avg_call_all = df[['x_id','length']].groupby(['x_id']).mean().sort_values(by='length',ascending=False)
count_calls_22 = df[['y_id','type']][df.x_id==22].groupby(['y_id']).count().sort_values(by='type',ascending=False)

display(top_calls_22)
display(avg_call_all)
display(count_calls_22)
```

```python
df[['x_id','y_id']]
```

```python
m_df= df_all[df_all.type=='message']
top_msgs_22 = m_df[['y_id','length','type']][m_df.x_id==22].groupby(
    ['y_id']).agg(sum_col1=('length', 'sum'),
                  count_col1=('length', 'count')).sort_values(by=['sum_col1'],ascending=False)
avg_msgs_all = m_df[['x_id','length']].groupby(['x_id']).mean().sort_values(by='length',ascending=False)
count_msgs_22 = m_df[['y_id','type']][m_df.x_id==22].groupby(['y_id']).count().sort_values(by='type',ascending=False)

display(top_msgs_22)
display(avg_msgs_all)
display(count_msgs_22)
```

```python
#Choose the x_id value

userx = 1

df_userx = df_all[df_all.x_id== userx]
df_userx_call = df_userx[df_userx.type=='call']
df_userx_text = df_userx[df_userx.type=='message']

#All Interactions
userx_interactions = len(df_userx)
userx_interactions_out = len(df_userx[df_userx.direction=='Outgoing'])
userx_init_all_pc = round((userx_interactions_out/userx_interactions)*100, 1)

#All Call related
userx_interactions_call = len(df_userx_call)
userx_interactions_text = len(df_userx_text)
userx_call_in = len(df_userx_call[df_userx_call.direction=='Incoming'])
userx_call_out = len(df_userx_call[df_userx_call.direction=='Outgoing'])
userx_call_miss = userx_interactions_call - (userx_call_in+userx_call_out)
userx_init_call_pc = round((userx_call_out/userx_interactions_call)*100, 1)

#All Text related
userx_text_in = len(df_userx_text[df_userx_text.direction=='Incoming'])
userx_text_out = len(df_userx_text[df_userx_text.direction=='Outgoing'])
userx_text_miss = userx_interactions_text - (userx_text_in+userx_text_out)
userx_init_text_pc = round((userx_text_out/userx_interactions_text)*100, 1)
```

```python
print(f'USER {userx} Interaction details')
display(userx_interactions)
display(userx_interactions_call) 
display(userx_interactions_text)
print(f'USER {userx} Call details')
display(userx_call_in)
display(userx_call_out)
display(userx_call_miss)
print(f'USER {userx} text details')
display(userx_text_in)
display(userx_text_out)
display(userx_text_miss)
```
